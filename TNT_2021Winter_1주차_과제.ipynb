# **Climate 데이터**

# **1번 문제**

```python
climate_data['date'] = pd.to_datetime(climate_data['date'])
```

```python
import seaborn as sns

fig=plt.figure()
x=climate_data['date']
co2=climate_data['co2']
rtemp=climate_data['relative_temp']

fig, ax1=plt.subplots()
ax1.set_xlabel('Time')
ax1.set_ylabel('CO2', color='blue')
ax1.plot(x, co2, color='blue')
ax1.tick_params(axis = 'y', colors = 'blue')

ax2=ax1.twinx()
ax2.set_xlabel('Time')
ax2.set_ylabel('Temp', color='red')
ax2.plot(x, rtemp, color='red')
ax2.tick_params(axis = 'y', colors = 'red')


plt.title('change of CO2 and Temperature', fontsize=10)
plt.show()
```

# **2번 문제**

```python
x=climate_data['date']
co2=climate_data['co2']
rtemp=climate_data['relative_temp']

fig, ax1=plt.subplots()
ax1.set_xlabel('Time')
ax1.set_ylabel('CO2')
ax1.plot(x, co2, color='blue')
plt.xticks(rotation = 90)
plt.title('chnage of CO2')

fig, ax2=plt.subplots()
ax2.set_xlabel('Time')
ax2.set_ylabel('relative temp')
ax2.plot(x, rtemp, color='red')
plt.xticks(rotation = 90)
plt.title('change of relative temp')
plt.show()
```

# **3번 문제**

```python
climate_data2 = climate_data[climate_data['date'] <= '1968-03-06']

fig, ax1=plt.subplots()
ax1.set_xlabel('Time')
ax1.set_ylabel('CO2')
plt.plot(climate_data2['date'], climate_data2['co2'], color = 'blue')
plt.xticks(rotation = 90)
plt.title('change of co2')

fig, ax2=plt.subplots()
ax2.set_xlabel('Time')
ax2.set_ylabel('relative temp')
plt.plot(climate_data2['date'], climate_data2['relative_temp'], color = 'red')
plt.xticks(rotation = 90)
plt.title('change of relative temp')

plt.tight_layout()
plt.show()
```

# **4번 문제**

```python
fig , ax2 = plt.subplots(figsize= (5,5))
plt.title('box plot of horsepower based on origin')
ax2=sns.boxplot(x='origin', y='horsepower', data=mpg_data, palette="Paired")
```

**Q. 그래프에서 얻을 수 있는 정보는?**

미국산 자동차의 마력이 평균이 가장 크고, 표준편차도 가장 크다.
일본산과 유럽산 자동차는 평균과 표준편차가 거의 비슷하지만, 유럽산 자동차에 이상치(Outlier)가 존재한다.

**Q. box plot에서 그래프 요소가 무엇을 의미하는지 작성해주세요**

최솟값 : 제 1사분위에서 1.5 IQR1을 뺀 위치이다. -> 상자 밖 가장 아래 선

제 1사분위(Q1) : 25%의 위치를 의미한다. -> 상자에서 아래쪽 선

제 2사분위(Q2) : 50%의 위치로 중앙값(median)을 의미한다. -> 상자 안 선

제 3사분위(Q3) : 75%의 위치를 의미한다. -> 상자에서 위쪽 선

최댓값 : 제 3사분위에서 1.5 IQR을 더한 위치이다. -> 상자 밖 가장 위 선

점 : 이상치
